## reading from csv file
- 1st with sql 
- 2nd with python

```python
%skip
%sql


CREATE TABLE bronze.crm_cust_info
USING DELTA
AS
SELECT * FROM read_files('/Volumes/workspace/bronze/raw_sources/source_crm/cust_info.csv', format => 'csv', header => true);
```

```python
df = (
    spark.read.option("header", "true")
    .option("inferSchema", "true")
    .csv("/Volumes/workspace/bronze/raw_sources/source_crm/cust_info.csv")
)

df.display()
```

## Write it to Bronze layer

```python
df.write.mode("overwrite").saveAsTable("bronze.crm_cust_info")
```
